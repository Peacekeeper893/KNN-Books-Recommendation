```python
import pandas as pd
import numpy as np
import pickle
```

### Load the model as well as the pivot matrix for reference.

```python
# load model_knn.pkl

with open('output/model_knn.pkl', 'rb') as f:
    model = pickle.load(f)

# import rating_popular_book_pivot
    
rating_popular_books = pd.read_csv('output/rating_popular_book.csv')
rating_popular_book_pivot = rating_popular_books.pivot(index='title', columns='user_id', values='rating').fillna(0)

rating_popular_book_pivot.head() 

```

### Query the model for recommenddations

```python
title = input("Enter the exact name of the book you want to get recommendations for: ")

if title not in rating_popular_book_pivot.index:
    # Find the closest Match
    import difflib
    matches = difflib.get_close_matches(title, rating_popular_book_pivot.index, n=1)

    try:
        if len(matches) > 0:
            title = matches[0]
        else:
            title = rating_popular_book_pivot.index[rating_popular_book_pivot.index.str.contains(title, case=False)].tolist()[0]
    except:
        print("Book not found. Please try again.")
        

print("You have entered: ", title)

```

```python

distances, indices = model.kneighbors(rating_popular_book_pivot.loc[title].values.reshape(1, -1), n_neighbors = 6)

for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(title))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, rating_popular_book_pivot.index[indices.flatten()[i]], distances.flatten()[i]))


```
